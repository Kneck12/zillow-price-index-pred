```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
guns = pd.read_csv('./../data/google_trend/google_terms_guns.csv')
```

```python
jobs = pd.read_csv('./../data/google_trend/google_terms_jobs.csv')
```

```python
lgbt = pd.read_csv('./../data/google_trend/google_terms_lgbt.csv')
```

```python
social = pd.read_csv('./../data/google_trend/google_terms_social.csv')
```

```python
travels = pd.read_csv('./../data/google_trend/google_terms_travels.csv')
```

```python
woty = pd.read_csv('./../data/google_trend/google_terms_woty.csv')
```

```python
disaster = pd.read_csv('./../data/google_trend/google_terms_disasters.csv')
```

```python
gentrification = pd.read_csv('./../data/google_trend/google_terms_gentrification2.csv')
```

```python
multifamily = pd.read_csv('./../data/google_trend/google_terms_multifamily_for_rent.csv')
```

```python
homes = pd.read_csv('./../data/google_trend/google_terms_homes_for_rent.csv')
```

```python
apartment = pd.read_csv('./../data/google_trend/google_terms_apartment_for_rent.csv')
```

```python
df_list = [guns, jobs, lgbt, social, travels, woty, disaster, gentrification, multifamily, homes, apartment]
i=0
ss = StandardScaler()
for item in df_list:
    item.columns = item.columns.map(lambda x: x.lower())
    item.drop(columns = item.columns[8:], inplace=True)
    for col in item.columns[3:]:
        temp = pd.DataFrame(item[col])
        temp = ss.fit_transform(temp)
        item[col] = temp
    item.drop(columns = 'ispartial', inplace = True)
    if i != 0:
        item.drop(columns = ['metro', 'datetime'], inplace = True)
    i+=1
```

```python
full_trends = pd.concat(df_list, axis = 1).fillna(0)
```
