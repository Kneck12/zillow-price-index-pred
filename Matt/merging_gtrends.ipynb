```python
apartment = pd.read_csv('./../data/google_trend/google_terms_apartment_for_rent.csv')
```

```python
df_list = [guns, jobs, lgbt, social, travels, woty, disaster, gentrification, multifamily, homes, apartment]
i=0
ss = StandardScaler()
for item in df_list:
    item.columns = item.columns.map(lambda x: x.lower())
    item.drop(columns = item.columns[8:], inplace=True)
    for col in item.columns[3:]:
        temp = pd.DataFrame(item[col])
        temp = ss.fit_transform(temp)
        item[col] = temp
    item.drop(columns = 'ispartial', inplace = True)
    if i != 0:
        item.drop(columns = ['metro', 'datetime'], inplace = True)
    i+=1
```

```python
full_trends = pd.concat(df_list, axis = 1).fillna(0)
```

```python
full_trends_backup = full_trends.copy()
```

```python
full_trends['starbucks2'] = full_trends['starbucks'].mean(axis=1)
```

```python
full_trends.drop(columns = 'starbucks', inplace = True)
```

```python
full_trends.rename(columns = {'starbucks2' : 'starbucks'}, inplace = True)
```

```python
full_trends.drop(columns = ['tender age shelter', 'tourists attraction', 'gun licence'], inplace = True)
```

```python
full_trends.rename(columns = {'metro':'Metro'}, inplace = True)
```

```python
full_trends['year_usable'] = full_trends['datetime'].map(lambda x: int(x[0:4])+1)
```

```python
full_trends['month'] = full_trends['datetime'].map(lambda x: int(x[5:7]))
```
